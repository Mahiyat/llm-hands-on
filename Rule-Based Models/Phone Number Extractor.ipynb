**Import Required Libraries**

```python
import spacy
import re
```



*   import spacy: Imports the spacy library, which is used for natural language processing.
*   import re: Imports the re library, which is used for regular expression operations.



**Load Spacy Model**

```python
nlp=spacy.load('en_core_web_sm')
```

nlp = spacy.load('en_core_web_sm'): Loads the small English language model from spacy. This model contains the necessary tools and data for processing English text, including tokenization, part-of-speech tagging, named entity recognition, and more.

**Define Regular Expression Pattern**

```python
pattern=r"\d{3}-\d{3}-\d{4}"
```

pattern = r"\d{3}-\d{3}-\d{4}": Defines a regular expression pattern to match a phone number in the format 123-456-7890.

*   \d{3}: Matches exactly three digits.
*   -: Matches a hyphen.
*   \d{3}: Matches exactly three digits.
*   -: Matches a hyphen.
*   \d{4}: Matches exactly four digits.





**Define Text to be Processed**

```python
text="Please call me at 123-456-7890"
```

text = "Please call me at 123-456-7890": Defines the text in which to search for the phone number.

**Process Text with Spacy**

```python
doc=nlp(text)
```

doc = nlp(text): Processes the text using the spacy model, creating a doc object that contains the processed text, including tokens, part-of-speech tags, entities, and more.

**Find Matches Using Regular Expression**

```python
for match in re.finditer(pattern,doc.text):
  start,end=match.span()
  span=doc.char_span(start,end)
  if span is not None:
    print("Found match:",span.text)
```



*   for match in re.finditer(pattern, doc.text): Uses re.finditer to find all matches of the pattern in the processed text (doc.text). re.finditer returns an iterator yielding match objects.
*   start, end = match.span(): For each match, gets the start and end positions of the match in the text.
*   span = doc.char_span(start, end): Creates a span object from the doc using the character start and end positions. This span object represents the matched text as a spacy Span, which retains all the linguistic annotations.
*   if span is not None: Checks if the span is not None (it could be None if the span does not align with token boundaries).
*   print("Found match:", span.text): Prints the text of the found match.



**Summary**<br>
This notebook processes a text using spacy to create a doc object. It then uses a regular expression to find phone numbers in the text. For each match, it creates a span object to represent the matched text as a spacy Span and prints the matched text. The regular expression pattern is designed to match phone numbers in the format 123-456-7890.
