**Define Function to Read Data**

```python
def read_data():
  data = ['This is a house', 'This is a home', 'I love my house', 'This is my home', 'Is this your house']
  dat = []
  for i in range(len(data)):
    for word in data[i].split():
      dat.append(word)
  print(dat)
  return dat
```



*   data: A list of strings (sentences).
*   dat: An empty list to store individual words.
*   The for loop iterates through each sentence in data and splits it into words, appending each word to dat.
*   The function prints and returns the list of words dat.



**Define Function to Create Bigrams and Count Frequencies**

```python
def create_bigram(data):
  list_of_bigrams = []
  bigram_counts = {}
  unigram_counts = {}
  l=len(data)-1
  for i in range((l+1)):
    if i<l and data[i+1].islower():
      list_of_bigrams.append((data[i],data[i+1]))
      if (data[i],data[i+1]) in bigram_counts:
        bigram_counts[(data[i], data[i+1])]+=1
      else:
        bigram_counts[(data[i], data[i+1])]=1
    if data[i] in unigram_counts:
      unigram_counts[data[i]]+=1
    else:
      unigram_counts[data[i]]=1
  return list_of_bigrams, unigram_counts, bigram_counts
```



*   list_of_bigrams: An empty list to store the bigrams.
*   bigram_counts: An empty dictionary to store the count of each bigram.
*   unigram_counts: An empty dictionary to store the count of each word (unigram).
*   The for loop iterates through the list data. If the current word is not the last word and the next word is lowercase, it forms a bigram and adds it to list_of_bigrams.
*   If the bigram is already in bigram_counts, its count is incremented. Otherwise, it is added with a count of 1.
*   Each word's count is tracked in unigram_counts.
*   The function returns list_of_bigrams, unigram_counts, and bigram_counts.



**Define Function to Calculate Bigram Probabilities**

```python
def calc_bigram_prob(list_of_bigrams, unigram_counts, bigram_counts):
  list_of_prob={}
  for bigram in list_of_bigrams:
    word1=bigram[0]
    word2=bigram[1]
    list_of_prob[bigram]=(bigram_counts.get(bigram))/(unigram_counts.get(word1))
  return list_of_prob
```



*   list_of_prob: An empty dictionary to store the probabilities of each bigram.
*   The for loop iterates through each bigram in list_of_bigrams. The probability of a bigram is calculated as the count of the bigram divided by the count of the first word in the bigram (using bigram_counts and unigram_counts).
*   The function returns the dictionary list_of_prob with the calculated probabilities.



**Execute the Functions and Print Results**

```python
data = read_data()
list_of_bigrams, unigram_counts, bigram_counts = create_bigram(data)
print("All possible Bigrams are:")
print(list_of_bigrams)
print("Bigrams along with their frequencies:")
print(bigram_counts)
print("Unigrams along with their frequencies:")
print(unigram_counts)
bigram_prob=calc_bigram_prob(list_of_bigrams, unigram_counts, bigram_counts)
print("Bigrams along with their probability:")
print(bigram_prob)
```



*   data = read_data(): Calls the read_data function to get the list of words.
*   list_of_bigrams, unigram_counts, bigram_counts = create_bigram(data): Calls the create_bigram function to get the bigrams and their counts.
*   The next few lines print the bigrams, their frequencies, and unigram frequencies.
*   bigram_prob = calc_bigram_prob(list_of_bigrams, unigram_counts, bigram_counts): Calls the calc_bigram_prob function to calculate and get the bigram probabilities.
*   The final print statement displays the bigrams along with their probabilities.
