**Import Required Libraries**

```python
import tensorflow
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model
```



*   import tensorflow: Imports the TensorFlow library.
*   import numpy as np: Imports the NumPy library for numerical operations.
*   from tensorflow.keras.models import Sequential: Imports the Sequential model class from Keras.
*   from tensorflow.keras.layers import Dense: Imports the Dense layer class from Keras.
*   from tensorflow.keras.utils import plot_model: Imports the plot_model utility from Keras for visualizing the model architecture.




**Define Training Data**

```python
train_data = np.array([[0,0], [0,1], [1,0], [1,1]])
train_labels = np.array([[1], [0], [0], [1]])
```



*   train_data: A NumPy array containing the training input data, representing a simple XNOR problem.
*   train_labels: A NumPy array containing the corresponding labels for the training data.



**Define the Model**

```python
model = Sequential([
    Dense(2, activation = 'relu', input_shape = (2,)),
    Dense(1, activation = 'sigmoid')
])
```

model = Sequential([...]): Defines a sequential model with two layers.


*   Dense(2, activation='relu', input_shape=(2,)): The first layer has 2 neurons with a ReLU activation function, and it expects input data with 2 features.
*   Dense(1, activation='sigmoid'): The second layer has 1 neuron with a sigmoid activation function, suitable for binary classification.



**Compile the Model**

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

model.compile(...): Compiles the model with the following parameters:


*   optimizer='adam': Uses the Adam optimizer.
*   loss='binary_crossentropy': Uses binary cross-entropy as the loss function.
*   metrics=['accuracy']: Tracks accuracy during training.



**Train the Model**

```python
model.fit(train_data, train_labels, epochs = 500, verbose = 0)
```

model.fit(...): Trains the model on the training data for 500 epochs. The verbose=0 argument suppresses the training output.

**Define Testing Data**

```python
test_data = np.array([[0,0], [0,1], [1,0], [1,1]])
test_label = np.array([[1], [0], [0], [1]])
```



*   test_data: A NumPy array containing the test input data.
*   test_label: A NumPy array containing the corresponding labels for the test data.



**Make Predictions**

```python
predictions = model.predict(test_data)
```

predictions = model.predict(test_data): Uses the trained model to make predictions on the test data.

**Evaluate the Model**

```python
for i in range(len(test_data)):
  print(f"Input: {test_data[i]} - Prediction: {np.round(predictions[i])} - Actual: {test_label[i]}")
```

The for loop iterates over each test sample, prints the input, the rounded prediction, and the actual label.

**Visualize the Model**

```python
plot_model(
    model,
    to_file='model.png',
    show_shapes=True,
    show_dtype=False,
    show_layer_names=True,
    rankdir='TB',
    expand_nested=False,
    dpi=200,
    show_layer_activations=True,
    show_trainable=False
)
```

plot_model(...): Generates a plot of the model architecture and saves it to a file named model.png. The options specified include:


*   show_shapes=True: Displays the shapes of the layers.
*   show_layer_names=True: Displays the names of the layers.
*   rankdir='TB': Sets the direction of the plot (top to bottom).
*   dpi=200: Sets the resolution of the plot.
*   show_layer_activations=True: Displays the activation functions used in each layer.




**Summary**<br>
This code defines a simple neural network using Keras, trains it on a small dataset representing an XNOR problem, and then makes predictions on a test dataset. The network consists of two layers: one hidden layer with 2 neurons and ReLU activation, and one output layer with 1 neuron and sigmoid activation. The model is trained using the Adam optimizer and binary cross-entropy loss, and the accuracy is tracked during training. Finally, the code visualizes the model architecture and saves it as an image.
