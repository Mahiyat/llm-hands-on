**Import Required Libraries**

```python
import spacy
from spacy import displacy
```



*   import spacy: Imports the spacy library, which is used for natural language processing tasks.
*   from spacy import displacy: Imports the displacy module from spacy, which is used for visualizing dependency parses and named entities.



**Load Spacy Model**

```python
nlp = spacy.load('en_core_web_sm')
```

nlp = spacy.load('en_core_web_sm'): Loads the small English language model from spacy. This model includes tools for tokenization, part-of-speech tagging, dependency parsing, named entity recognition, and more.

**Process the Text**

```python
doc = nlp("John saw a flashy blue hat at the store")
```

doc = nlp("John saw a flashy blue hat at the store"): Processes the input sentence using the spacy model, creating a doc object that contains the processed text, including tokens, part-of-speech tags, entities, dependencies, and more.

**Visualize Dependency Parse**

```python
displacy.render(doc, style = "dep")
```

displacy.render(doc, style="dep"): Uses the displacy module to render a visual representation of the dependency parse of the sentence. The style="dep" argument specifies that the visualization should show the dependencies (syntactic structure) of the sentence.

**Summary**<br>
This notebook processes a sentence to analyze its syntactic structure and uses displacy to visualize the dependency parse. The visualization will display the words in the sentence along with arrows indicating the grammatical relationships between them. For the sentence "John saw a flashy blue hat at the store", the visualization will show how the words are related to each other in terms of their syntactic roles, such as subject, object, and modifiers.

N.B. displacy.render(doc, style="dep") has been used as it is a Jupyter notebook. If you are not working with a Jupyter notebook, use displacy.serve(doc, style="dep").
