**Import Required Libraries**

```python
import spacy
import pandas as pd
```



*   import spacy: Imports the spacy library, which is used for natural language processing tasks including tokenization, part-of-speech tagging, and named entity recognition.
*   import pandas as pd: Imports the pandas library, a powerful data manipulation and analysis tool. Note that pandas is imported but not used in this specific code snippet.



**Load Spacy Model**

```python
nlp = spacy.load('en_core_web_sm')
```

nlp = spacy.load('en_core_web_sm'): Loads the small English language model from spacy. This model contains the necessary tools and data for processing English text, including tokenization, part-of-speech tagging, named entity recognition, and more.

**Define Sentence to be Processed**

```python
sentence = "Michael Jackson came to India in 1996 for a concert in Mumbai."
```

sentence = "Michael Jackson came to India in 1996 for a concert in Mumbai.": Defines the sentence in which to identify named entities.

**Process Sentence with Spacy**

```python
doc = nlp(sentence)
```

doc = nlp(sentence): Processes the sentence using the spacy model, creating a doc object that contains the processed text, including tokens, part-of-speech tags, entities, and more.

**Extract and Print Named Entities**

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```



*   for ent in doc.ents: Iterates over the named entities identified in the doc object. doc.ents is a list of Span objects representing the named entities.
*   print(ent.text, ent.label_): For each entity, prints the text of the entity (ent.text) and its label (ent.label_), which indicates the type of the entity (e.g., PERSON, GPE, DATE).



**Summary**<br>
This notebook processes a sentence using the spacy library to identify named entities. It loads the pre-trained small English model, processes the input sentence to create a doc object, and then iterates over the identified entities, printing each entity's text and label.
